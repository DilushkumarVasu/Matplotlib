# Histogram

```python
import numpy as np

# Generate 10,000 random numbers
X = np.random.randn(10000)

# Check the minimum and maximum values
min_value = np.min(X)
max_value = np.max(X)

print(f'Min value: {min_value}, Max value: {max_value}')
```

```python
import numpy as np

# Define different sample sizes
sample_sizes = [100, 1000, 10000]

for size in sample_sizes:
    X = np.random.randn(size)
    min_value = np.min(X)
    max_value = np.max(X)
    print(f'Sample size: {size} => Min value: {min_value}, Max value: {max_value}')
```

# Matplotlib Exercise

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.uniform(-1,1,(1000,2))
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1]);
```

```python
#xor logical representation method1
Y = np.zeros(1000)
Y[(X[:, 0] < 0) & (X[:, 1] > 0)] = 1  # (-x, +y)
Y[(X[:, 0] > 0) & (X[:, 1] < 0)] = 1  # (+x, -y)
plt.scatter(X[:,0],X[:,1],c = Y);
```

```python
#xor logical representation method2
# Define colors based on the XOR condition
Y = np.zeros(1000)
Y[(X[:, 0] * X[:, 1]) > 0] = 1  # Same sign for x and y -> Color 1 (e.g., Blue)
plt.scatter(X[:,0],X[:,1],c = Y);
```

```python
#xor logical representation method3
def make_xor_map(x, y, N):
  colorv = np.zeros(N)
  for i in range(N):
    if x[i] * y[i] < 0:
      colorv[i] = 1
  return colorv

N = 1000
X = np.random.rand(N, 2) * 2 - 1
Y = make_xor_map(X[:,0], X[:, 1], N)
plt.scatter(X[:,0], X[:, 1], c= Y, cmap='bwr', alpha=0.5)
plt.show()
```
